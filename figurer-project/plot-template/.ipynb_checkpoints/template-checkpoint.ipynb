```python
#example exporting datas (plotting template is separate below)
import numpy as np

x = np.linspace(0,100,1000)
y= np.sin(x)

#exporting data
np.savetxt('data1.txt', np.transpose([x,y]))

y=np.cos(x)**2
#exporting data
np.savetxt('data2.txt', np.transpose([x,y]))


np.savez('data.npz',x=x,y=np.tan(x))


print('saved datas cell finished')
```

```python
#template for plot back from data

#imports
import matplotlib.pyplot as plt
import numpy as np


def global_decorations():
    #global decorations code pasted here
    plt.legend(['generic legend'])


def figurer():
    plt.figure()
    global_decorations()

    
### BEGIN FIGURE
#load data (do whenever data save time is changed)
data1=np.loadtxt('data1.txt')

#only one call to figure allowed. one plt.figure() == one \includegraphics in latex
figurer()
plt.plot(data1[:,0],data1[:,1])
plt.legend(['specific legend'])
plt.ylabel('y')
plt.xlabel('x')
plt.axis([0,100,-2,2])
plt.title('data1.txt')
### END FIGURE



### BEGIN FIGURE

data2=np.loadtxt('data2.txt')

plt.figure()
plt.plot(data2[:,0],data2[:,1])
plt.ylabel('y')
plt.xlabel('x')
plt.axis([20,60,-.1,1.22])
plt.title('data2.txt')
### END FIGURE

### BEGIN FIGURE

#could load more than once
npzdata = np.load('data.npz')
data2=np.loadtxt('data2.txt')


plt.figure()
plt.plot(npzdata['x'],npzdata['y'])
plt.ylabel('y')
plt.xlabel('x')
plt.axis([20,60,-.1,1.22])
plt.title('data.npz with data2.txt')
plt.plot(data2[:,0],data2[:,1])
### END FIGURE
```
